```python
import pandas as pd
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt

# Carregando os conjuntos de dados
anime_df = pd.read_csv('anime.csv')
rating_df = pd.read_csv('rating.csv')

# Tratando valores ausentes, se necessário
anime_df.fillna(value={'genre': 'Unknown', 'type': 'Unknown', 'episodes': 0}, inplace=True)
rating_df.fillna(value={'rating': -1}, inplace=True)

# Filtrando as avaliações superiores a 0
rating_df = rating_df[rating_df['rating'] > 0]

# Juntando os conjuntos de dados usando a coluna anime_id
merged_df = pd.merge(rating_df, anime_df, on='anime_id')

# Dividindo os dados em conjunto de treinamento e teste
train_data, test_data = train_test_split(merged_df, test_size=0.2, random_state=42)

# Exibindo as colunas presentes nos conjuntos de treinamento e teste
print("Colunas no conjunto de treinamento:", train_data.columns)
print("\nColunas no conjunto de teste:", test_data.columns)

# Selecionando as features e o target
features = ['user_id', 'anime_id']
target = 'rating'

# Criando e treinando o modelo MLPRegressor
model = MLPRegressor(hidden_layer_sizes=(100, 50), max_iter=500, random_state=42)

try:
    # Tentando ajustar o modelo com base nas features e no target
    model.fit(train_data[features], train_data[target])

    # Realizando previsões no conjunto de teste
    predictions = model.predict(test_data[features])

    # Calculando a raiz do erro quadrático médio (RMSE)
    rmse = sqrt(mean_squared_error(test_data[target], predictions))
    print(f'RMSE: {rmse}')
except KeyError as e:
    print(f"Erro: {e}")
```

```python
import pandas as pd
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt

# Carregando os conjuntos de dados
anime_df = pd.read_csv('anime.csv')
rating_df = pd.read_csv('rating.csv')

# Tratando valores ausentes, se necessário
anime_df.fillna(value={'genre': 'Unknown', 'type': 'Unknown', 'episodes': 0}, inplace=True)
rating_df.fillna(value={'rating': -1}, inplace=True)

# Filtrando as avaliações superiores a 0
rating_df = rating_df[rating_df['rating'] > 0]

# Reduzindo o tamanho dos conjuntos de dados para treinamento e teste
rating_df = rating_df.sample(frac=0.1, random_state=42)
merged_df = pd.merge(rating_df, anime_df, on='anime_id')

# Dividindo os dados em conjunto de treinamento e teste
train_data, test_data = train_test_split(merged_df, test_size=0.2, random_state=42)

# Exibindo as colunas presentes nos conjuntos de treinamento e teste
print("Colunas no conjunto de treinamento:", train_data.columns)
print("\nColunas no conjunto de teste:", test_data.columns)

# Selecionando as features e o target
features = ['user_id', 'anime_id']
target = 'rating_x'  # Corrigindo para 'rating_x'

# Reduzindo a complexidade do modelo
model = MLPRegressor(hidden_layer_sizes=(50,), max_iter=200, random_state=42)

try:
    # Tentando ajustar o modelo com base nas features e no target
    model.fit(train_data[features], train_data[target])

    # Realizando previsões no conjunto de teste
    predictions = model.predict(test_data[features])

    # Calculando a raiz do erro quadrático médio (RMSE)
    rmse = sqrt(mean_squared_error(test_data[target], predictions))
    print(f'RMSE: {rmse}')
except KeyError as e:
    print(f"Erro: {e}")
```

```python
# Função para o Feedback Bandit
def feedback_bandit(user_id, anime_ids):
    ratings_predicted = model.predict(pd.DataFrame({'user_id': user_id, 'anime_id': anime_ids}))
    return ratings_predicted

# Exemplo de como usar a função
user_id_example = 1
anime_ids_example = [123, 456, 789]

predicted_ratings = feedback_bandit(user_id_example, anime_ids_example)
print("Predicted Ratings:", predicted_ratings)
```

```python
import pandas as pd
from sklearn.neural_network import MLPRegressor
import numpy as np

# Função para o Feedback Bandit com Epsilon-Greedy
def feedback_bandit_epsilon_greedy(user_id, anime_ids, epsilon=0.1):
    # Com probabilidade epsilon, escolhe uma ação aleatória
    if np.random.rand() < epsilon:
        random_ratings = np.random.uniform(0, 10, len(anime_ids))
        return random_ratings
    else:
        # Com probabilidade 1-epsilon, escolhe a ação com a maior previsão de avaliação
        ratings_predicted = model.predict(pd.DataFrame({'user_id': user_id, 'anime_id': anime_ids}))
        return ratings_predicted

# Exemplo de como usar a função com Epsilon-Greedy
user_id_example = 1
anime_ids_example = [123, 456, 789]

epsilon = 0.1  # Defina o valor de epsilon conforme necessário
predicted_ratings = feedback_bandit_epsilon_greedy(user_id_example, anime_ids_example, epsilon)
print("Predicted Ratings (Epsilon-Greedy):", predicted_ratings)
```
